```python
import processing_data
import pretty_printer
```

```python
# Pour décoder les dates en français
import locale
locale.setlocale(locale.LC_TIME, "fr_FR")
```

```python
pipeline = processing_data.Pipeline()


calendar = pipeline.add_step(
    processing_data.transformer.JSONReader(
        processing_data.data.school_holidays(), 'Calendrier')
).add_step(
    processing_data.transformer.Date(dict(Debut="%Y-%m-%d",
                                          Fin="%Y-%m-%d",
                                          DateDebut="%A %d %B %Y",
                                          DateFin="%A %d %B %Y"))
).add_step(
    processing_data.transformer.Select(dict(Zone="Zone C"))

).add_step(
    processing_data.transformer.Select(dict(annee_scolaire="2020-2021"))
).run()
pretty_printer.print(calendar)
```

```python
pipeline.clear


academy = pipeline.add_step(
    processing_data.transformer.JSONReader(
        processing_data.data.school_holidays(), 'Academie')
).add_step(
    processing_data.transformer.Select(dict(Zone="Zone C"))
).run()
pretty_printer.print(academy)
```

```python
table = academy.join(calendar, on="Zone")
```

```python
pretty_printer.print(table)
```
